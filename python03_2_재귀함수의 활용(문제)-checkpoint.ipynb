#재귀함수(Recursive Function)
+ 재귀함수란 함수에서 자신을 다시 호출하여 작업을 수행하는 방식의 함수
-  함수 정의 내에 같은 이름의 함수가 오는 경우 재귀함수라고 함.
- 재귀함수 사용시 반드시 탈출 조건이 있어야 stack overflow를 방지할 수 있다.




###[실습]1. 팩토리얼이란 특정 정수에서 1까지의 수를 모두 곱하는 것을 말한다. 정수가 입력되면 반복문을 사용하여 팩토리얼 함수를 구현하는 코드를 작성하시오.

[입출력]  
```
정수입력 : 6  
6 5 4 3 2 1   
6! = 720
```

````python
data = int(input('정수입력 : '))

def fact(n):
    res = 1
    for i in range(n, 0, -1):
      print(i, end=' ')
      res *= i
    return res
# 함수를 작성하시오.

print(f'\n{data}! = {fact(data)}')
````

###[실습]2. 위의 팩토리얼 함수를 재귀함수로 구현하시오.

````python
data = int(input('정수입력 : '))
def fact2(n):
  print(n, end=' ')
  if n == 1:
    return 1
  return n * fact2(n-1)
 # 재귀함수를 작성하시오.
print(f'\n{data}! = {fact2(data)}')
````

###[실습]3.재귀함수를 이용하여 어떤 10진수 n이 주어지면 2진수로 변환해서 출력하시오.

```
[입력예시1]
11
[출력예시1]
1011

[입력예시2]
0
[출력예시2]
0
```

###[실습]4. 피보나치 수열이란 첫째 항, 둘째 항이 1이고 그 뒤의 항은 바로 앞의 두 항의 합인 수열을 뜻한다.
피보나치 수열 : 1, 1, 2, 3, 5, 8, 13, 21 ····
재귀함수로 피보나치 수열을 구현하고 연산 수행 횟수를 출력해 보자.(그림출처 : https://chanos.tistory.com/entry/)  
image.png
```
[입출력]
피보나치 항 입력 : 35
피보나치 35항 : 9227465
계산에 활용된 연산 횟수 : 18454929
```

````python
# 피보나치 함수 코드를 재귀함수로 작성하고 항의 값을 출력해 보자.
count = 0
def fibo(n) :
  global count
  count += 1
  if n <= 2 :
    return 1
  else :
    return fibo(n-1) + fibo(n-2)


a = int(input('피보나치 항 입력 : '))
print(f'피보나치 {a}항 : {fibo(a)}')
print(f'계산에 활용된 연산 횟수 : {count}')
````

###[실습]5. 연산횟수를 줄이기 위해 피보나치 수열 문제를 메모제이션을 이용해 구현해 보자.
fibo(n)의 값을 구하기 위해서는 fibo(n-1)와 fibo(n-2)의 값을 알아야 한다. 재귀는  n-1과 n-2의 경우의 값을 저장해두고 연산 횟수를 줄일 수 있는데, 이러한 논리를 활용한 기법이 메모이제이션(Memoization)이다.
위 예제에서 메모제이션을 사용하여 피보나치 수열을 구현하고, 연산횟수를 출력해 보자.  
조건1) 딕셔너리를 사용하여 연산한 항을 저장한다.  
조건2) 실행횟수를 구하시오.
```
[입출력]  
몇 번째 피보나치 수를 계산할까요? 10
10번째 피보나치 수는 55입니다.
총 연산횟수 : 17
{1: 1, 2: 1, 3: 2, 4: 3, 5: 5, 6: 8, 7: 13, 8: 21, 9: 34, 10: 55}
```

````python
#메모할 딕셔너리
# 기말고사
memo = {}

count = 0
def fibo(n):
  global count
  count += 1
  if n <= 2:
    return 1
  if n in memo:
    return memo[n]
  memo[n] = fibo(n-1) + fibo(n-2)
  return memo[n]
#재귀함수를 작성하시오.

n = int(input("몇 번째 피보나치 수를 계산할까요? "))
# 결과 출력
print(f"{n}번째 피보나치 수는 {fibo(n)}입니다.")
print(f"총 연산횟수 : {count}")

````

>**중첩된 리스트가 있을때 중첩을 모두 제거하고 1차원 리스트로 만드는것:**



###[실습]6. 리스트의 요소를 순회하면서 요소가 리스트이면 재귀함수를 이용하여 리스트 요소에 리스트를 타입을 풀어서 1차원 리스트로 만드시오.

````python
#리스트 평탄화 코드 작성
example = [[1,2,3],[4,[5,6]],7,[8,9]]
def flatten(data) :
  print(f"flatten({data})")
  output = []
  for item in data:
    if type(item) == list:
      output += flatten(item)
    else:
      output.append(item)
  return output
  #코드를 작성하시오.


print(flatten(example))
````

###[해결]2. 주어진 양의 정수 n에 대해 1부터 n까지의 모든 숫자의 합을 계산하는 재귀함수를 작성하시오.
```
[입력]
10
[출력]
55
```

````python
#코드를 작성하시오.
n = int(input())
def sum(n):
  if n == 1:
    return 1
  return n + sum(n - 1)
print(sum(n))
````

###[해결]3. 문자열을 입력받아 해당 문자열을 거꾸로 출력하는 재귀함수를 작성하시오.
```
[입력]
hello
[출력]
olleh
```


````python
 #코드를 작성하시오.
k = input().strip()
def reverse(s):
  if len(s) == 0:
    return s
  return s[-1] + reverse(s[:-1])
print(reverse(k))
````

###[해결]4. 문자열이 대칭인지(회문인지) 확인하는 재귀함수를 작성하시오.
```
[입력1]
level
[출력1]
회문입니다.
[입력2]
racecar
[출력2]
회문이 아닙니다.
```

````python
 #코드를 작성하시오.
s = input().strip()
def f(left, right):
  if left >= right:
    return True

  if s[left] == s[right]:
    return f(left + 1, right - 1)
  else:
    return False
if f(0, len(s) - 1):
  print("회문입니다.")
else:
  print("회문이 아닙니다.")
````

#TypeError: 'int' object is not callable

<에러 발생 원인>

우선 해당 오류가 발생한 이유는 예약어를 변수명으로 사용 하였기 때문이다. 그리고 기초 프로그래밍을 공부하며 이러한 실수를 많이 할 수 있는 예약어들이 sum(), min(), max() 등이 있을 것이다.

​

이러한 예약어들은 각각의 기능들이 있는 함수이다. (이름만 봐도 어떤 기능인지는 짐작이 간다.)

기초 프로그래밍을 하다보면 합을 구하는 문제, 최댓값, 최솟값 등을 구하는 문제들이 많이 나오는데 이 경우, 위와 같은 단어들을 변수명으로 사용하는 경우가 종종 있을 것이다.

​

하지만 이미 (예를들어) sum을 변수명으로 사용한 후에 sum() 함수를 다시 사용할 경우 이름의 중복으로 인해 해당 오류가 발생하게 된다.

​

<해결법>

결론적으로 del을 이용하여 오류를 만든 변수를 삭제해주면 된다.

ex) del sum
