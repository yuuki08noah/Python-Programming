### 10.  입력과 출력
* 사용자로부터 두 개의 숫자를 입력받은 후 두 개의 숫자를 더한 값, 곱한 값, 나눈 값, 나눈 몫, 나머지 값을 각각 다음과 같이 출력하는 프로그램을 작성하세요.

Example<br>
INPUT: 4 and 4 <br>
OUTPUT: <br>

4 + 4 = 8 <br>
4 * 4 = 16 <br>
4 / 4 = 1 <br>
4 // 4 = 1 <br>
4 % 4 = 0 <br>

```python
x, y = map(int, input().split())
print (f'{x} + {y} = {x+y}')
print (f'{x} * {y} = {x*y}')
print (f'{x} / {y} = {x/y}')
print (f'{x} // {y} = {x//y}')
print (f'{x} % {y} = {x%y}')
```

### 11.  입력과 출력
* 사용자로부터 두 개의 숫자를 입력받은 후 두 개의 숫자를 더한 값, 곱한 값, 나눈 값, 나눈 몫, 나머지 값을 각각 다음과 같이 출력하는 프로그램을 작성하세요.
  - 단, 나눈 값은 소숫점 첫번째 자리까지만 출력하세요.

Example<br>
INPUT: 5 and 4 <br>
OUTPUT: <br>

5 + 4 = 9 <br>
5 + 4 = 20 <br>
5 / 4 = 1.2 <br>
5 // 4 = 1 <br>
5 % 4 = 1 <br>

```python
x, y = map(int, input().split())
print (f'{x} + {y} = {x+y}')
print (f'{x} * {y} = {x*y}')
print (f'{x} / {y} = {x/y:.1f}')
print (f'{x} // {y} = {x//y}')
print (f'{x} % {y} = {x%y}')
```
